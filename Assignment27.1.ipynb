### 1) How-to-count-distance-to-the-previous-zero For each value, count the difference back to the previous zero   (or the start of the Series, whichever is closer) create a new column 'Y'

```python
import pandas as pd
df = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]} )
Y = []
cnt = 0
for i in df.X: 
    if i == 0: 
        cnt = 0;
    Y.append(cnt)
    cnt+=1
df['Y']= Y
print(df)
```

### 2) Create a DatetimeIndex that contains each business day of 2015 and use it to index a Series of random numbers.

```python
import pandas as pd 
import random  
bd = pd.bdate_range('2015-01-01', '2015-12-31', freq='B')
data_nums = []
for i in bd: data_nums.append(random.random())
df = pd.DataFrame({'Rndm_num': data_nums}, index = bd )
df
```

### 3) Find the sum of the values in s for every Wednesday

```python
wed_sum= df.groupby(df.index.day_name()).sum()['Wednesday':'Wednesday'].values[0][0]
print("The sum of Wednesday values is.....",wed_sum)
```

### 4) Average For each calendar month

```python
df.groupby(df.index.month_name()).mean()
```
